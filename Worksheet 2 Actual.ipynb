```python
import matplotlib.pyplot as plt 
import numpy as np
import math

fm = 0.02

def x(t1):
    return [math.cos(2*math.pi*fm*i) for i in t1]
    
t = np.arange(-51.0, 51.0, 0.01)

xorg = x(t)

fs1 = 0.02
fs2 = 0.04
fs3 = 0.5

ts = 50

ts1 = 1/ fs1
ts2 = 1/ fs2
ts3 = 1/ fs3

n1 = np.arange((-51.0*fs1), (51.0*fs1), 1)
n2 = np.arange((-51.0*fs2), (51.0*fs2), 1)
n3 = np.arange((-51.0*fs3), (51.0*fs3), 1)

x1 = x((n1*ts1))
x2 = x((n2*ts2))
x3 = x((n3*ts3))

plt.figure(figsize = [10,5])
plt.subplot(221)
plt.plot(t, xorg)
plt.title('Given Signal x(t)')
plt.xlabel('t')
plt.ylabel('amplitude')
plt.grid()
plt.show()

plt.figure(figsize = [10,5])
plt.subplot(222)
plt.title('x1[n] [fs = 0.02]')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.stem(n1, x1)
plt.grid()
plt.show()

plt.figure(figsize = [10,5])
plt.subplot(223)
plt.stem(n2, x2)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('x2[n] [fs = 0.04]')
plt.grid()
plt.show()

plt.figure(figsize = [10,5])
plt.subplot(224)
plt.stem(n3, x3)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('x3[n] [fs = 0.5]')
plt.grid()
plt.show()
```
