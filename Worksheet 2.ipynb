```python
import matplotlib.pyplot as plt 
import math
import numpy as np

%matplotlib inline
x=[]
y=[]
def prog_plot(i):
    if(i>=-1 and i<0):
        return (2*i)
    if(i>=0 and i<5):
        return (math.exp(-i)+math.sin(2*math.pi*i))
    else:
        return 0
def signal(i):
    x=prog_plot(i)
    if type(x)==None:
        return 0
    else:
        return x
i=-5
while(i>=-5 and i<=5):
    x.append((i))
    y.append(prog_plot(i))
    i=i+0.1
plt.figure(figsize=(20,10))
plt.figure(1)
plt.subplot(4,1,1)
plt.title('Operations on x(t)')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.legend('x(t)')
plt.grid()
plt.plot(x,y)
plt.show
#x(t/2 +0.5)
i=-5
x=[]
y=[]
while(i>=-5 and i<=5):
    x.append((i))
    y.append(prog_plot(i/2 + 0.5))
    i=i+0.1
plt.figure(figsize=(20,10))
plt.figure(2)
plt.subplot(4,1,2)
plt.title('Custom Signal x(t/2+0.5)')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.legend('x(t/2+0.5)')
plt.plot(x,y)
plt.grid()
plt.show
#even part of the signal 
i=-5
x=[]
y=[]
while (i>=-5 and i<=5):
    x.append(i)
    y.append((signal(i)+signal(-i))/2)
    i=i+0.01
plt.figure(figsize=(20,10))
plt.figure(3)
plt.subplot(4,1,3)
plt.title('Even Component of x(t)')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.legend('even signal')
plt.plot(x,y)
plt.grid()
plt.show
#odd part of the signal
i=-5
x=[]
y=[]
while (i>=-5 and i<=5):
    x.append(i)
    y.append((signal(i)-signal(-i))/2)
    i=i+0.01
plt.figure(figsize=(20,10))
plt.figure(4)
plt.subplot(4,1,4)
plt.title('Odd Component of x(t)')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.legend('odd signal')
plt.plot(x,y)
plt.grid()
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
i=0
x=[]
y=[]
while(i>=0 and i<=10):
    x.append(i)
    y.append((8*math.pow(3,i)))
    i=i+1
plt.figure(1)
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.legend('h[n]')
plt.title('Impulse Response h[n]')
plt.stem(x,y)
plt.grid()
plt.show
#frequency response
import cmath
i=0
x=[]
y=[]
j=cmath.sqrt(-1)
while(i>=0 and i<4):
    x.append(i)
    l1=8/(math.cos(i)-(j*math.sin(i))-(3))
    y.append((abs(l1)))
    i=i+0.1
plt.figure(2)
plt.figure(figsize=(20,10))
plt.subplot(2,1,2)
plt.title('Frequency Response H[w]')
plt.xlabel('w')
plt.ylabel('H(w)')
plt.legend('H(w)')
plt.plot(x,y)
plt.grid()
plt.show
```

```python
import matplotlib.pyplot as plt
import math 

%matplotlib inline
x=[]
y=[]
i=-4
while(i>=-4 and i<=4):
    if(i>=-3 and i<=-1):
        x.append(i)
        y.append(math.exp(-0.5*i))
    if (i>=-1 and i<=0):
        x.append(i)
        y.append((-3*i)+1)
    if (i>=0 and i<=2):
        x.append(i)
        y.append(1)
    if(i>=2 and i<=4):
        x.append(i)
        y.append(math.exp(0.5*i)*(math.sin(5*math.pi*i)))
    i=i+0.1
plt.figure(1);
plt.figure(figsize=(15,7))
plt.title("Custom Signal x[t]")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.plot(x,y)
plt.grid()
plt.show
```
