```
import numpy as np
import matplotlib.pyplot as plt

def circular_ops(x, n, r, k):
  if r == 0:
    try:
      x = [x[i-k] for i in range(0, len(n))]
    except IndexError:
      x = [x[-len(n)+i-k] for i in range(0, len(n))]
  elif r == 1:
    x = [x[-(i-k)] for i in range(0, len(n))]
  return np.array(x)

def zeroes(n):
  a = [0 for i in range(0,n)]
  return np.array(a)

def concat(x,y):
  b = []
  for i in range(0, len(x)):
    b.append(x[i])
  for j in range(0, len(y)):
    b.append(y[j])
  return np.array(b)
```

```
n = np.arange(0, 8, 1)
N = 8

xu = [1,2,3,-2,6,4]
x = concat(xu, zeroes(N-len(xu)))
print(x)

x_fold = circular_ops(x, n, 1, 0)
print(x_fold)

plt.figure(1, figsize = [10,5])
plt.title('Given Signal with Zero Padding x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid()
plt.stem(n, x)

plt.figure(2, figsize = [10,5])
plt.title('Circularly Folded Signal x[-n]')
plt.xlabel('n')
plt.ylabel('x[-n]')
plt.grid()
plt.stem(n,x_fold)
```

```
#Perform circular convolution using DFT
#Check similarity to linear convolution

xu = [5,4,3,2,2,4,6,8]
hu = [4,3,2,1]

l1 = len(xu)
l2 = len(hu)

x = concat(xu, zeroes((l1+l2-1) - l1))
h = concat(hu, zeroes((l1+l2-1) - l2))

n = np.arange(0,l1 + l2 -1,1)

X = np.fft.fft(x)
H = np.fft.fft(h)
Y = X * H

y = np.fft.ifft(Y)

plt.figure(15, figsize=[10,5])
plt.grid()
plt.title('Convolved Output y[n] using Circular Convolution')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.stem(n,y)

y1 = np.convolve(xu,hu)

plt.figure(16, figsize=[10,5])
plt.grid()
plt.title('Convolved Output y[n] using Linear Convolution')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.stem(n,y1)
```

```
def sumSquares(x):
  sumx = 0
  for i in x:
    sumx = sumx + (np.abs(i)**2)
  return sumx

x = [1,5,3,7,-20,12,8,2]
N = len(x)

sum1 = sumSquares(x)
print("Energy in the time domain: ",sum1,"J")

X = np.fft.fft(x)
sum2 = sumSquares(X)/N 
print("Energy in the frequency domain: ",sum2,"J")
```
