```
import numpy as np
import matplotlib.pyplot as plt 
from math import *

t = np.arange(-0.2, 0.3, 0.01)
#pi = math.pi

x = np.sin(2*pi*5*t)

fs = 50
Ts = 1/fs
n = np.arange(-0.2*fs, 0.2*fs, 1)

xn = np.sin(2*pi*5*Ts*n)

N = len(n)
X = np.fft.fft(xn, N)
k = np.arange(0, N, 1)

fa = (k/N) * fs
fd = fa/ fs

plt.figure(1)
#plt.figure(figsize=[20,10])
#plt.subplot(411)
plt.plot(t, x)
plt.grid()
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Given Signal x(t)')

plt.figure(2)
#plt.figure(figsize=[20,10])
#plt.subplot(412)
plt.stem(n, xn)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('Sampled Signal x[n]')
plt.grid()

plt.figure(3)
#plt.figure(figsize=[20,10])
#plt.subplot(413)
plt.stem(k, np.abs(X))
plt.xlabel('k')
plt.ylabel('|X[k]|')
plt.title('Magnitude Spectrum |X[k]|')
plt.grid()

plt.figure(4)
#plt.figure(figsize=[20,10])
#plt.subplot(414)
plt.stem(k, np.angle(X))
plt.xlabel('k')
plt.ylabel('<X[k]')
plt.title('Phase Spectrum <X[k]')
plt.grid()

plt.figure(5)
#plt.figure(figsize=[20,10])
#plt.subplot(2,1,1)
plt.stem(fa, np.abs(X))
plt.xlabel('fa')
plt.ylabel('|X[k]|')
plt.grid()
plt.title('|x[k]| vs fa')

plt.figure(6)
#plt.figure(figsize=[20,10])
#plt.subplot(2,1,2)
plt.stem(fd, np.abs(X))
plt.xlabel('fd')
plt.ylabel('|X[k]|')
plt.grid()
plt.title('|X[k]| vs fd')
```

```
import numpy as np
import matplotlib.pyplot as plt 
from math import *

f1 = 5
f2 = 10
f3 = 20
 
fs = 100
Ts = 1/fs
 
t = np.arange(-0.2, 0.2, 0.0001)
n = np.arange(-0.2*fs, 0.2*fs, 1)
 
x1 = np.sin(2*pi*f1*t)
x2 = np.sin(2*pi*f2*t)
x3 = np.sin(2*pi*f3*t)
 
x1n = np.sin(2*pi*f1*n*Ts)
x2n = np.sin(2*pi*f2*n*Ts)
x3n = np.sin(2*pi*f3*n*Ts)
 
xn = x1n + x2n + x3n
 
N = len(n)
 
x = x1 + x2 + x3
X = np.fft.fft(x,N)
 
k = np.arange(0, N, 1)
 
plt.figure(1)
plt.plot(t, x)
plt.title('Given Signal x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid()

plt.figure(2)
plt.stem(n, xn)
plt.title('Sampled Signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid()

plt.figure(3)
plt.stem(k, np.abs(X))
plt.title('Magnitude Spectrum |X[k]|')
plt.xlabel('k')
plt.ylabel('|X[k]|')
plt.grid()
 
plt.figure(4)
plt.stem(k, np.angle(X))
plt.title('Phase Spectrum <X[k]')
plt.xlabel('k')
plt.ylabel('<X[k]')
plt.grid()
```

```
import numpy as np
import matplotlib.pyplot as plt 
from math import *

def concat(x, y):
  l = [i for i in x]
  for j in y:
    l.append(j)
  return np.array(l)

f1 = 5
f2 = 10
f3 = 20
 
fs = 100
Ts = 1/fs
 
t1 = np.arange(0, 0.4, 0.0001)
t2 = np.arange(0.4, 0.8, 0.0001)
t3 = np.arange(0.8, 0.12, 0.0001)

n1 = np.arange(0, 0.4*fs, 1)
n2 = np.arange(0.4*fs, 0.8*fs, 1)
n3 = np.arange(0.8*fs, 0.12*fs, 1)
 
x1 = np.sin(2*pi*f1*t1)
x2 = np.sin(2*pi*f2*t2)
x3 = np.sin(2*pi*f3*t3)

x = concat(concat(x1,x2),x3)

x1n = np.sin(2*pi*f1*n1*Ts)
x2n = np.sin(2*pi*f2*n2*Ts)
x3n = np.sin(2*pi*f3*n3*Ts)

xn = concat(concat(x1n,x2n),x3n)

nn = concat(concat(n1,n2),n3)
t = concat(concat(t1,t2),t3)

N = len(n1) + len(n2) + len(n3)

X = np.fft.fft(xn, N)

k = np.arange(0, N, 1)

plt.figure(1, figsize=[10,5])
plt.plot(t, x)
plt.grid()
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Given Signal x(t)')
plt.show()

plt.figure(2, figsize=[10,5])
plt.grid()
plt.stem(nn, xn)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('Sampled Signal x[n]')
plt.show()

plt.figure(3, figsize=[10,5])
plt.stem(k, np.abs(X))
plt.grid()
plt.xlabel('k')
plt.ylabel('|X[k]|')
plt.title('Magnitude Spectrum of X[k]')
plt.show()

plt.figure(4, figsize=[10,5])
plt.stem(k, np.angle(X))
plt.grid()
plt.xlabel('k')
plt.ylabel('<X[k]')
plt.title('Phase Spectrum of X[k]')
plt.show()
```
